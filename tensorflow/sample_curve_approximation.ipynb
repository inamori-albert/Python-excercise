```python
import tensorflow as tf
import numpy as np
import time

import matplotlib.pyplot as plt
%matplotlib inline
```

```python
# 1次関数の近似
start = time.time()

x_data = np.random.rand(100).astype("float32")
y_data = 0.1 * x_data + 0.3

W = tf.Variable(tf.random_uniform([1], -1.0, 1.0))
b = tf.Variable(tf.zeros([1]))
y = W * x_data + b

loss = tf.reduce_mean(tf.square(y_data - y))
optimizer = tf.train.GradientDescentOptimizer(0.5)
train = optimizer.minimize(loss)

init = tf.initialize_all_variables()

sess = tf.Session()
sess.run(init)

for step in range(1001):
    sess.run(train)
    if step % 100 == 0:
        print(step, sess.run(W), sess.run(b))

pred_x_data = []
pred_y_data = []
for i in range(100):
    print(i, sess.run(W) * i + sess.run(b))
    pred_x_data.append(i/100.0)
    pred_y_data.append(sess.run(W) * (i/100.0) + sess.run(b))

sess.close()

timer = time.time() - start
print(("time:{0}".format(timer)) + "[sec]")
```

```python
# 結果の確認
fig = plt.figure()
ax1 = fig.add_subplot(1,1,1)
# 元の結果
ax1.scatter(x_data, y_data, color='blue')
# 予測結果
plt.scatter(pred_x_data, pred_y_data, color='red')
fig.show()
```
