```python
import time
import numpy as np
import pandas as pd

import tensorflow as tf

import matplotlib.pyplot as plt
%matplotlib inline
```

```python
# データのインポート
from tensorflow.examples.tutorials.mnist import input_data
mnist = input_data.read_data_sets("MNIST_data/", one_hot=True)
```

```python
# データの確認
image_list = np.reshape(mnist.train.images[0],(28,28))

# imageをファイルに出力
# pd.DataFrame(image_list).to_csv("sample.csv")

# ラベルを表示
print(mnist.train.labels[0])
```

```python
# 手書き文字認識の学習

start = time.time()

# モデルの作成
x = tf.placeholder(tf.float32,[None,784]) # 入力のプレースホルダー
W = tf.Variable(tf.zeros([784,10])) # モデルの重み
b = tf.Variable(tf.zeros([10])) # バイアス
y = tf.nn.softmax(tf.matmul(x,W)+b)

# 損失コスト関数とオプティマイザーを定義
y_ = tf.placeholder(tf.float32,[None,10]) # 正解のプレースホルダー
cross_entropy = -tf.reduce_sum(y_*tf.log(y)) # 損失コスト関数：交差エントロピー
train_step = tf.train.GradientDescentOptimizer(0.01).minimize(cross_entropy) # 学習率0.01で、勾配降下アルゴリズムを使って交差エントロピーを最小化


# 作成した変数の初期化
init = tf.initialize_all_variables()
# sessionのlaunch
sess = tf.Session()
sess.run(init)
# 訓練
for i in range(1000):
   batch_xs, batch_ys = mnist.train.next_batch(100) # 100のランダムなデータのバッチを取得し訓練
   sess.run(train_step, feed_dict={x: batch_xs, y_: batch_ys})
   print(tf.argmax(y, 1))

# 訓練モデルのテスト
correct_prediction = tf.equal(tf.argmax(y,1),tf.argmax(y_,1))
accuracy = tf.reduce_mean(tf.cast(correct_prediction, "float"))
print(sess.run(accuracy, feed_dict={x: mnist.test.images, y_: mnist.test.labels}))

timer = time.time() - start
print(("time:{0}".format(timer)) + "[sec]")
```
